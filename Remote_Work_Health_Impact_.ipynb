```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv('post_pandemic_remote_work_health_impact_2025.csv')
df.head()
df.info()
df.columns
df.isnull().sum()
df['Work_Arrangement'].value_counts().plot(kind='bar', title="Work Arrangement Distribution")
plt.ylabel("Count")
plt.show()

pd.crosstab(df['Work_Arrangement'], df['Mental_Health_Status']).plot(kind='bar', stacked=True)
pd.crosstab(df['Work_Arrangement'], df['Burnout_Level']).plot(kind='bar', stacked=True)
df.groupby('Work_Arrangement')[['Work_Life_Balance_Score', 'Social_Isolation_Score']].mean().plot(kind='bar')
pd.crosstab(df['Work_Arrangement'], df['Salary_Range']).plot(kind='bar', stacked=True, figsize=(10,6))
pd.crosstab(df['Work_Arrangement'], df['Physical_Health_Issues']).plot(kind='bar', stacked=True)


```
